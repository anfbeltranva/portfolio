```python
df_groupGeo=corriente.groupby(['Departamento'],as_index=False).mean()
df_groupGeo
```

```python
response = json.loads(requests.get("https://gist.githubusercontent.com/john-guerra/43c7656821069d00dcbc/raw/3aadedf47badbdac823b00dbe259f6bc6d9e1899/colombia.geo.json")._content)
```

```python

colombia_map=folium.Map(location=[4.1156735, -72.9301367], zoom_start=6)

folium.Choropleth(
    geo_data=response,
    data=df_groupGeo,
    columns=['Departamento','Precio'],
    key_on='feature.properties.NOMBRE_DPT',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Precios por departamento'
    
    ).add_to(colombia_map)


colombia_map

```

```python
xserie=np.arange(1,len(grupoFecha2)+1)
```

```python
xdata=xserie/max(xserie)
ydata=grupoFecha2['Precio']/max(grupoFecha2['Precio'])
```

```python
msk = np.random.rand(len(grupoFecha2)) < 0.8
train_x = xdata[msk]
test_x = xdata[~msk]
train_y = ydata[msk]
test_y = ydata[~msk]
```

```python
plt.figure(figsize=(15,8))
plt.plot(train_x,train_y, 'ro')
plt.plot(test_x,test_y, 'bo')
plt.ylabel('Precio')
plt.xlabel('tiempo')
plt.show()
```

```python
reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
```

```python
models, predictions = reg.fit(train_x,test_x,train_y,test_y)
```
